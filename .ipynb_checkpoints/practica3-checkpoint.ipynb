```python
"""
En este ejercicio utilizamos los datos de gapminder (recordar el ejercicio 10 de la Practica 2
para acceder a los datos de gapminder).
(a) Utilizando seaborn.objects, graficar la curva de la expectativa de vida en Argentina en
funcion del a˜no, completando el siguiente codigo. Sugerencia: recordar de la practica
anterior como filtrar datos de un dataset.
"""
from gapminder import gapminder
import seaborn . objects as so

datosArgentina = gapminder[gapminder.country == 'Argentina']
#print(datosArgentina)

(
so . Plot ( data = datosArgentina , x = "year", y = "lifeExp")
. add ( so . Line () )
)
```

```python
"""
(b) Realizar un nuevo grafico donde puedan verse las curvas de la expectativa de vida de los
paıses de America en funcion del a˜no, una curva por cada paıs.
Sugerencia: utilizar los parametros group = ??? o color = ???. ¿Cual es la diferencia
entre los dos?
"""
americas = gapminder[gapminder.continent == 'Americas']
#display(americas)
(
    so.Plot(data = americas, x = 'year', y = 'lifeExp')
    . add ( so . Line (), color = 'country' )
)
#se diferencia en la visibilidad en este caso
```

```python

"""
(c) Queremos agregar al grafico del ıtem anterior una curva de tendencia lineal utilizando el
metodo .add(so.Line(), so.PolyFit(1)). ¿Cual de las siguientes dos formas de agrupar
los datos es la forma correcta? Explicar la diferencia entre los dos codigos.
"""

# Codigo 1
(
so . Plot ( data = americas , x ='year' , y ='lifeExp' , group = 'country')
. add ( so . Lines ( color ="#bbca") )
. add ( so . Line () , so . PolyFit (1) )
)
# Codigo 2
(
so . Plot ( data = americas , x ='year' , y ='lifeExp' , group = 'country')
. add ( so . Lines ( color ="#bbca") , group = 'country')
. add ( so . Line () , so . PolyFit (1) )
)

```

```python
"""
(d) Realizar el siguiente grafico, con las curvas de expectativa de vida agrupadas por continente. Sugerencias: ¿que hace el metodo facet() de seaborn.objects.Plot()? ¿Y el
parametro wrap = ??? de facet()?
"""

#facet() permite hacer una plantilla de varios graficos, "wrap=" permite controlar la cantidad de columnas de gráficos mostradas
(
    so.Plot(data = gapminder, x="year", y="lifeExp")
    . add ( so . Lines ( color ="#bbca") , group = 'country')
    .add(so . Line () ,so.PolyFit(1))
    .facet(col="continent", wrap=3)
)
```

```python
"""
3. Como vimos en el Ejercicio 1c, si asignamos una codificacion (o mapeo) al definir un Plot(),
el mapeo se asigna en todas las capas de marcas (objetos mark). En cambio, si asignamos una
codificacion dentro del metodo add() de una marca, mapeo se realiza solo en esa capa. Por
ultimo, si asignamos un parametro de la marca, el valor se asigna directamente (ver grafico).
"""

"""
¿Que resultado esperan para el siguiente grafico? ¿Cuales codificaciones se pasan de Plot()
a Dot() y cuales no pueden pasarse? ¿Cuales codificaciones se establecen en Dot()? ¿Cuales
variables estan asignadas directamente en Dot()? ¿De que color van a pintarse los puntos?
"""


"""
Resultado esperado del gráfico:

Se espera un gráfico de dispersión donde el eje x representa la longitud del pico en milímetros (bill_length_mm) y el eje y representa la profundidad del pico en milímetros (bill_depth_mm).
Los puntos en el gráfico estarán marcados por especie (species).
El color de los puntos indicará la especie.
El grosor de los bordes de los puntos indicará la masa corporal (body_mass_g).
Los puntos tendrán un color de relleno de .8 (un tono de gris).
Codificaciones pasadas de Plot() a Dot():

Las codificaciones pasadas de Plot() a Dot() son: x, y, edgewidth, marker, linestyle, y color.
Codificaciones que no pueden pasarse de Plot() a Dot():

La codificación que no puede pasarse de Plot() a Dot() es edgecolor.
Codificaciones establecidas en Dot():

La codificación establecida en Dot() es color, que se establece en el parámetro color=".8".
Variables asignadas directamente en Dot():

Ninguna variable se asigna directamente en Dot() en este caso.
Color de los puntos:

Los puntos se pintarán con un tono de gris, ya que el color de relleno se establece en .8.
"""
display(
    so . Plot (
    penguins , x ="bill_length_mm", y ="bill_depth_mm",
    edgewidth ="body_mass_g", marker = "species",
    linestyle = "island", color = "species"
    )
.add(so.Dot( color =".8"), edgecolor ="sex")

)
```

```python
"""
(b) Usando la funcion facet() separar el grafico del item anterior en tres subgraficos, uno
para cada especie.
"""
display(
    so.Plot(
    penguins, x = 'bill_depth_mm', y = 'bill_length_mm', color= 'sex')
    .label(x="Depth", y="Length", color="")
    
    .add(so.Dot()
    ).facet(col="species", wrap=1)
)
```

```python
"""
(b) Realizar un grafico de barras con la cantidad de ping¨unos en cada isla, completando el
siguiente codigo
"""

pinguinos_por_isla = penguins.groupby('island').size() # Usar el codigo del item anterior .

(
so . Plot ( x = pinguinos_por_isla.index, y = pinguinos_por_isla)
. add ( so . Bar () )
)
```

```python
"""
(c) El grafico que acabamos de hacer es un histograma categorico (usamos una variable
categorica en el eje X). Podemos realizar el mismo grafico usando la funcion Hist()
para contar automaticamente las cantidades (sin definir una variable pinguinos_por_isla.
Completar el siguiente codigo.
"""
(
so . Plot ( data= penguins , x= penguins['island'], y=penguins.index)
. add ( so . Bars () , so . Hist () )
)
```
