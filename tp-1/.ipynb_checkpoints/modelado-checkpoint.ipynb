a) Generar un DataFrame en el que las columnas sean las líneas de colectivo (TIPO_TRANSPORTE
== "COLECTIVO") de AMBA de jurisdicción nacional (PROVINCIA == "JN") y las observaciones sean los días del año. Es decir, cada fila del DataFrame corresponde a un día
del año, y en esa fila deben figurar la cantidad de pasajeros que viajaron en cada línea de
colectivo en el día correspondiente.

```python
datos_ColectivoJN = df[df['TIPO_TRANSPORTE'] == 'COLECTIVO']
datos_ColectivoJN = df[df['PROVINCIA'] == 'JN']
datos_ColectivoJN.head()
```

```python
datos_ColectivoJN.info()
```

```python
cols = datos_ColectivoJN.LINEA.unique()
pasajeros_por_linea = pd.DataFrame()

for col in cols:
    datos_linea = datos_ColectivoJN[datos_ColectivoJN.LINEA == col ][["DIA_TRANSPORTE", "CANTIDAD"]]
    datos_linea = datos_linea.set_index("DIA_TRANSPORTE").rename(columns = {"CANTIDAD" : col })
    pasajeros_por_linea = pd.concat ([pasajeros_por_linea, datos_linea ], axis = 1)

pasajeros_por_linea.head()
```

```python
pasajeros_por_linea.info()
```

b) Eliminar las columnas correspondientes a líneas de colectivo que tengan datos faltantes.

```python
# Crear una lista de columnas a mantener
columns_to_keep = ['DIA_TRANSPORTE']

# Identificar columnas con datos vacíos (NaN)
columns_with_na = pasajeros_por_linea.columns[pasajeros_por_linea.isna().any()].tolist()
print(f'Cantiad de lineas de colectivo que van a ser removidas: {len(columns_with_na)}')


# Filtrar las columnas que no son 'DIA_TRANSPORTE'
columns_to_drop = [col for col in columns_with_na if col not in columns_to_keep]

# Eliminar las columnas con datos vacíos excepto 'DIA_TRANSPORTE'
pasajeros_por_linea = pasajeros_por_linea.drop(columns=columns_to_drop)

# Chequeo
columns_with_na = pasajeros_por_linea.columns[pasajeros_por_linea.isna().any()].tolist()
print(f'Lineas con filas vacias: {len(columns_with_na)}')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.model_selection import train_test_split, learning_curve, cross_val_score, cross_validate, KFold
from sklearn.linear_model import LinearRegression, SGDRegressor

import warnings
warnings.filterwarnings("ignore")
```

Preparamos los datos de entrenamiento y testeo.

```python
lineas = ['BS_ASLINEA_123',	'BSAS_LINEA_002', 'BSAS_LINEA_009',	'BSAS_LINEA_010', 'BSAS_LINEA_015']
x = pasajeros_por_linea[lineas] 
y = pasajeros_por_linea['BSAS_LINEA_009'] 
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

Vamos a hacer una primera prueba en donde usamos el Regresor Lineal y lo entrenamos sin usar validacion cruzada y despues con para
comparar la eficacicia de usar este enfoque mencionado. Vamos a usar la metrica de MSE para medir los resultados del modelo.

```python
from sklearn.linear_model import Ridge
Ridge(alpha=2).get_params()
```

```python
# Enfoque 'Naive' sin CV
lr = LinearRegression() # r = Ridge(alpha=2)
lr.fit(X_train, y_train)

#Test Set
y_pred_test = lr.predict(X_test)
error_test = metrics.mean_squared_error(y_test, y_pred_test)

print(f"Test MSE Error {error_test}")
```

Analizemos entrenando con validacion cruzada con los distintos metodos que nos ofrece scikit-learn.

```python
help(cross_val_score)
```

```python
# Entrenar el modelo con validación cruzada y obtener los puntajes
for alpha in [1,2,3]:
    cv_scores[alpha] = cross_val_score(Ridge(alpha=alpha), X_train, y_train, scoring='neg_mean_squared_error') # 5 CV siendo ejecutados por default

# Convertir los puntajes a positivos para interpretabilidad
cv_scores = -cv_scores

# Mostrar los puntajes de validación cruzada
print(f'Puntajes de validación cruzada (MSE): {cv_scores}')
print(f'Promedio de los puntajes de validación cruzada (MSE): {np.mean(cv_scores)}')
```

```python
scores = cross_validate(lr, X_train, y_train, scoring="neg_mean_squared_error", return_train_score=True)
scores
```

```python
# Convertir los puntajes a positivos para interpretabilidad
cv_scores = -scores['test_score']

# Mostrar los puntajes de validación cruzada
print(f'Puntajes de validación cruzada (MSE): {cv_scores}')
print(f'Promedio de los puntajes de validación cruzada (MSE): {np.mean(cv_scores)}')
```

Podemos observar como por unicamente haber entrenado, el mismo modelo, con la tecnica mencionada pudimos
reducir en 2 i-esimos el valor de error.

```python
# Calcular las curvas de aprendizaje
train_sizes, train_scores, test_scores = learning_curve(lr, X_train, y_train, cv=5, scoring='neg_mean_squared_error')
```

```python
print(train_sizes) # Como quedaron dividadas los 5 subset 
print(train_scores*(-1)) 
print(test_scores*(-1))
```

```python
for train_size, cv_train_scores, cv_test_scores in zip(
    train_sizes, train_scores, test_scores
):
    print(f"{train_size} samples were used to train the model:")
    print(f"The MSE value is {cv_train_scores.mean()*(-1)}")
    print(f"The MSE is {cv_test_scores.mean()*(-1)}\n")
```

Podemos ver como se van calculando para cada uno de los 5 subsets que se van generando, cada vez mas grandes,
el valor de MSE en para cada set llegando al mismo valor que con los otros metodos demostrados.

Podemos notar la utilidad de este ultimo metodo para poder observar tambien como fue evolucionando, no solo 
el valor del MSE para los set de validacion, sino tambien para los de entrenamiento. Ademas, tenemos ya todos
los valores disponibles para visualizar graficamente el comportamiento del entrenamiento con las curvas de 
aprendizaje,

```python
train_scores_mean = -np.mean(train_scores, axis=1)
test_scores_mean = -np.mean(test_scores, axis=1)

# Visualizar las curvas de aprendizaje
plt.figure(figsize=(12, 6))
plt.plot(train_sizes, train_scores_mean, 'o-', color='r', label='Training error')
plt.plot(train_sizes, test_scores_mean, 'o-', color='g', label='Validation error')

# Añadir etiquetas y leyenda
plt.title('Curvas de Aprendizaje')
plt.xlabel('Tamaño del conjunto de entrenamiento')
plt.ylabel('Error cuadrático medio (MSE)')
plt.legend(loc='best')
plt.grid()
plt.show()
```
