```python
X = macro_oferta[[col for col in macro_oferta.columns if (col.endswith(".oferta")) & ~(col.startswith('PBI'))]]
y = macro_oferta[[col for col in macro_oferta.columns if col.startswith('PBI')]]
```

(c) Ajustar un modelo de regresi´on lineal ordinaria o Ridge y generar el vector de predicciones (hacerlo sobre el conjunto de todos los datos, no es necesario separar en entrenamiento y testeo).

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV



ridge_model = Ridge()

alphas = np.array([0.001, 0.005, 0.01, 0.02, 0.1, 0.5, 1])
grid = GridSearchCV(estimator = ridge_model, param_grid=dict(alpha=alphas))
grid.fit(X,y)

best_alpha = grid.best_estimator_.alpha
best_alpha
```

```python
ridge_model = Ridge(alpha=best_alpha)

ridge_model.fit(X,y)

predicciones = ridge_model.predict(X)
```
