#### (b) En el dataset penguins, convertir todos los nombres de variables a mayusculas utilizando listas por comprension.

```python
penguins_mayus_pc = penguins.copy()
penguins_mayus_pc.columns = [str.upper(variable) for variable in penguins_mayus_pc.columns]
penguins_mayus_pc.head()
```

#### En el dataset penguins, convertir solo los nombres de variables que empiezan con bill a mayusculas.

#### Sugerencia: Si queremos definir una lista for comprension aplicando distintas funciones podemos crear una funcion partida utilizando if / else. 

#### Por ejemplo, ¿cual sera la salida de la siguiente instruccion?

#### [x * 10 if x % 2 == 0 else x for x in [1,2,3,4,5,6]]

#### Nota: la funcion que utilizamos para definir listas por comprension es una funcion lambda (pero no necesitamos escribir lambda). Esta sintaxis para definir funciones partidas es parte de la sintaxis de las funciones lambda.

```python
penguins_mayus_bill = penguins.copy()
penguins_mayus_bill.columns = [str.upper(variable) if variable.startswith('bill') else variable for variable in penguins_mayus_bill.columns]
penguins_mayus_bill.head()
```

#### 8. Utilizando una funci´on lambda, agregar una nueva columna a la base de datos llamada peso_bin que contenga:

#### “chico” si la masa corporal es menos que 4000 gramos.

#### “grande” si la masa corporal es mayor que 4000 gramos.

```python
penguins_peso_bin = penguins_notnan.copy()
penguins_peso_bin['peso_bin'] = penguins_peso_bin['body_mass_g'].apply(lambda x: 'chico' if x < 4000 else 'grande')
display(penguins_peso_bin.head())
display(penguins_peso_bin.tail())
print("Le saque los nan porque sino me devolvia grande cuando no tengo datos")
```

#### 11. Empezando con penguins realizar las siguientes operaciones:

(a) Crear una nueva variable que tenga el peso en kg.

(b) Convertir la variable island a min´uscula. Sugerencia: aplicar .str.upper() a la columna

```python
penguins_11 = penguins.copy()
penguins_11['peso_kg'] = penguins_11['body_mass_g'].apply(lambda x: x/1000)
penguins_11.head()
```

```python
# Voy a suponer que la guia esta mal y quisieron decir que todos los valores sean mayuscula
penguins_11['island'] = penguins_11['island'].apply(lambda x: str.upper(x))
penguins_11.head()
```

#### 12. Agregar una columna a penguins con la mediana de la masa corporal de los ping¨uinos de cada especie usando group_by() y agg().

```python
penguins_12 = penguins.copy()
# penguins_12.describe()
medianas_especie = penguins_12.groupby('species')['body_mass_g'].agg('median')

# Fusiona la información de las medianas por especie con el DataFrame original
penguins_12 = penguins_12.merge(medianas_especie.rename('mediana_especie'), on='species', how='left')

medianas_especie.head()
```
