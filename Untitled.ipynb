```python
import pandas as pd
import seaborn as sns
penguins = sns.load_dataset("penguins")
penguins.columns
```

```python
penguins_adelie_prom = penguins[penguins['species'] == 'Adelie'].body_mass_g.mean()
penguins_adelie_prom
```

```python
penguins_adelie__Dream_prom = penguins[(penguins['species'] == 'Adelie') & (penguins['island'] == 'Dream')].body_mass_g.mean()
penguins_adelie__Dream_prom
```

```python
penguins_islands = penguins.groupby('island')['body_mass_g'].std()
penguins_islands
```
