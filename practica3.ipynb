```python
"""
En este ejercicio utilizamos los datos de gapminder (recordar el ejercicio 10 de la Pr´actica 2
para acceder a los datos de gapminder).
(a) Utilizando seaborn.objects, graficar la curva de la expectativa de vida en Argentina en
funci´on del a˜no, completando el siguiente c´odigo. Sugerencia: recordar de la pr´actica
anterior como filtrar datos de un dataset.
"""
from gapminder import gapminder
import seaborn . objects as so

datosArgentina = gapminder[gapminder.country == 'Argentina']
#print(datosArgentina)

(
so . Plot ( data = datosArgentina , x = "year", y = "lifeExp")
. add ( so . Line () )
)
```

```python
"""
(b) Realizar un nuevo gr´afico donde puedan verse las curvas de la expectativa de vida de los
pa´ıses de Am´erica en funci´on del a˜no, una curva por cada pa´ıs.
Sugerencia: utilizar los par´ametros group = ??? o color = ???. ¿Cu´al es la diferencia
entre los dos?
"""
americas = gapminder[gapminder.continent == 'Americas']
#display(americas)
(
    so.Plot(data = americas, x = 'year', y = 'lifeExp')
    . add ( so . Line (), color = 'country' )
)
#se diferencia en la visibilidad en este caso
```

```python

"""
(c) Queremos agregar al gr´afico del ´ıtem anterior una curva de tendencia lineal utilizando el
m´etodo .add(so.Line(), so.PolyFit(1)). ¿Cu´al de las siguientes dos formas de agrupar
los datos es la forma correcta? Explicar la diferencia entre los dos c´odigos.
"""

# Codigo 1
(
so . Plot ( data = americas , x ='year' , y ='lifeExp' , group = 'country')
. add ( so . Lines ( color ="#bbca") )
. add ( so . Line () , so . PolyFit (1) )
)
# Codigo 2
(
so . Plot ( data = americas , x ='year' , y ='lifeExp' , group = 'country')
. add ( so . Lines ( color ="#bbca") , group = 'country')
. add ( so . Line () , so . PolyFit (1) )
)

```

```python
"""
(d) Realizar el siguiente gr´afico, con las curvas de expectativa de vida agrupadas por continente. Sugerencias: ¿qu´e hace el m´etodo facet() de seaborn.objects.Plot()? ¿Y el
par´ametro wrap = ??? de facet()?
"""

#facet() permite hacer una plantilla de varios graficos, "wrap=" permite controlar la cantidad de columnas de gráficos mostradas
(
    so.Plot(data = gapminder, x = 'year', y = 'lifeExp', group='continent')
    .add(so . Line (), color='country')
).facet(wrap=5)

## FALTA TERMINAR
```
