```python
import scipy
import numpy as np
import pandas as pd
import seaborn as sns
import seaborn.objects as so

# Para clustering
from sklearn.datasets import make_blobs
from sklearn.datasets import make_circles
from sklearn.preprocessing import StandardScaler 
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
```

```python
# Tomamos nubes aleatorias y graficamos los puntos.
X, v = make_blobs(n_samples=60, centers=4, cluster_std=0.60, random_state=0)
datos = pd.DataFrame(X)
datos.columns = ["x", "y"]
datos.head()
```

```python
# Elegimos eps mas chico que la distancia que vemos entre clusters.
# Elegimos min_samples segn la densidad de puntos (cuntos puntos tienen a distancia menor que eps los puntos del cluster?)
clustering = DBSCAN(eps=1, min_samples=6)
etiqueta = clustering.fit_predict(X)
print(etiqueta) # el -1 es un outlier
```

La salida de la funcin debe ser tipo "set" (conjunto), que permite fcilmente incorporar nuevos elementos al conjunto sin repetir.

```python
# Datos
X, labels_true = make_circles(n_samples=1000, factor=0.5, noise=0.05, random_state = 42)
X = StandardScaler().fit_transform(X)

datos = pd.DataFrame(X)
datos.columns = ["x", "y"]

# Graficamos
(
    so.Plot(data = datos, x = "x", y = "y")
    .add(so.Dot())
)
```

```python
# Pintamos el punto 999
ind = 999
etiquetas = np.zeros(len(datos))
etiquetas[ind] = 1

(
    so.Plot(data = datos, x = "x", y = "y", color = etiquetas.astype("str"))
    .add(so.Dot())
)
```

```python
def obtener_vecinos(datos, ind_punto, eps):
    punto = datos.iloc[ind_punto]
    dist = np.sqrt(np.sum((datos-punto)**2, axis = 1))  # Lo vimos la clase pasada
    vecinos = datos[dist<eps].index
    return(vecinos)
```

```python
cjto = obtener_vecinos(datos, ind, 0.1)
print(cjto)
```

**Ejercicio 2**

1. Implementar una funcin que dado un conjunto de puntos (en un DataFrame), un vector de etiquetas correspondientes a clusters y la etiqueta de algn cluster, agregue al cluster todos los puntos directamente alcanzables desde algn punto central del cluster.
2. Aplicar la funcin al ejemplo del ejercicio anterior.
3. Graficar todos los puntos, pintando de un color distinto los puntos hallados en el punto 2.

Repetir los puntos 2 y 3 un par de veces y observar como crece el cluster. Qu faltara para poder encontrar el cluster completo?

```python
def extender_cluster(datos, etiquetas, cluster, eps, minPts):
    indices_cluster = datos[etiquetas == cluster].index
    for ind_punto in indices_cluster:
        vecinos = obtener_vecinos(datos, ind_punto, eps)
        if (len(vecinos) >= minPts):
            etiquetas[list(vecinos)] = cluster

    print("Puntos en el cluster: ", np.sum(etiquetas == cluster))
    return(etiquetas)            
```

```python
# Pintamos el punto 999
ind = 999
etiquetas = np.zeros(len(datos))
etiquetas[ind] = 1

eps = 0.3
minPts = 5
```

```python
etiquetas = extender_cluster(datos, etiquetas, etiquetas[ind], eps, minPts)
(
    so.Plot(datos, x = "x", y = "y", color = etiquetas.astype("str"))
    .add(so.Dot())
)
```

**Ejercicio 3**
1. Realizar mediante un for 20 iteraciones del procedimiento anterior y graficar el cluster resultante.
2. Si la cantidad de iteraciones no alcanz para calcular todo el cluster aumentar la cantidad de iteraciones.
3. Identificar algn punto que no est en el cluster encontrado, calcular el cluster correspondiente y graficar.

```python
eps = 0.2
minPts = 5
ind = 999
etiquetas = np.zeros(len(datos))
etiquetas[ind] = 1

for i in range(20):
    etiquetas = extender_cluster(datos, etiquetas, 1, eps, minPts)
    
(
    so.Plot(datos, x = "x", y = "y", color = etiquetas.astype("str"))
    .add(so.Dot())
)
```

```python
# Ahora queremos construir el otro cluster
etiquetas[0] = 2

for i in range(40):
    etiquetas = extender_cluster(datos, etiquetas, 2, eps, minPts)
    
(
    so.Plot(datos, x = "x", y = "y", color = etiquetas.astype("str"))
    .add(so.Dot())
)
```

```python
eps = 0.2
minPts = 5
ind = 999
etiquetas = np.zeros(len(datos))
etiquetas[ind] = 1

while True:
    etiquetasViejas = etiquetas.copy()
    etiquetas = extender_cluster(datos, etiquetas, 1, eps, minPts)
    if np.all(etiquetas == etiquetasViejas):
        break
(
    so.Plot(datos, x = "x", y = "y", color = etiquetas.astype("str"))
    .add(so.Dot())
)
```

# Ejemplo: deteccin de centros de actividad urbana
Fuente: https://bitsandbricks.github.io/post/dbscan-machine-learning-para-detectar-centros-de-actividad-urbana/

La S del nombre DBSCAN se refiere a "SPATIAL", este algoritmo es especialmente til para detectar clusters en informacin espacial o en el plano, por ejemplo barrios, comunidades, focos de tormentas, etc.

Vamos a utilizarlo para detectar focos gastronmicos en la ciudad de Mendoza (ms particularmente, bares).

```python
# Cargamos los datos
df = pd.read_csv("../Datos/mendoza_poi.csv")
df
```

```python
# Tenemos negocios de distintas categorias
df.categoria.unique()
```

```python
# Veamos los distintos tipos de locales gastronomicos

# Utilizamos query en lugar de df[df.categoria == "gastronomia"]
df.query("categoria == 'gastronomia'").tipo.unique()
```

```python
# Nos quedamos solo con los bares
datosBares = df.query("categoria == 'gastronomia' and tipo == 'bar'").reset_index()
datosBares
```

```python
# Graficamos
so.Plot(data = datosBares, x = "lng", y = "lat").add(so.Dot())
```

```python
# Aplicamos DBSCAN eligiendo los parmetros arbitrariamente.
# eps: a que distancia esperamos que esten los bares cercanos
# minPts: cuantos bares esperamos que tenga cerca un bar para considerarlo una zona de bares

# Usamos solo las variables de latitud y longitud para el agrupamiento.
# Para puntos ms alejados (datos de un pas o continente), podemos usar una funcin auxiliar para calcular la distancia entre puntos)

clustering = DBSCAN(eps=0.01, min_samples=6)
etiqueta = clustering.fit_predict(datosBares[["lng", "lat"]])

(
    so.Plot(data = datosBares, x = "lng", y = "lat", color = etiqueta.astype("str"))
    .add(so.Dot())
)
```

# Seleccin del hiperparmetro eps

Los valores eps y minPts son hiperparmetros, no se pueden aprender de los datos, pero algunas tcnicas nos pueden ayudar.

Para el valor de esp, calculamos para cada bar a qu distancia est el bar ms cercano y graficamos los valores ordenados de menor a mayor.

```python
from sklearn.neighbors import NearestNeighbors

dataset = datosBares[["lng", "lat"]]
neighbors = NearestNeighbors(n_neighbors=2)  # Esta funcin nos devuelve los ms cercanos incluyendo a si mismo, por eso tomamos 2.
neighbors_fit = neighbors.fit(dataset)
```

```python
# Ordenamos de menor a mayor las distancias y graficamos
distances, indices = neighbors_fit.kneighbors(dataset)
distances = distances[:,1]
distances = np.sort(distances, axis=0)

so.Plot(x = np.arange(len(distances)), y = distances).add(so.Line())
```

```python
# Elegimos el eps donde la curva hace un codo.
# De esta forma tenemos un valor pequeo que incluye a la mayoria de los datos.
# Tomamos eps = 0.005.
clustering = DBSCAN(eps=0.005, min_samples=6)
etiqueta = clustering.fit_predict(datosBares[["lat", "lng"]])

(
    so.Plot(data = datosBares, x = "lat", y = "lng", color = etiqueta.astype("str"))
    .add(so.Dot())
)
```

Entontramos 6 zonas de bares, una grande central y otras perifericas. 

Si queremos separar la zona central en distintas zonas, podemos tomar un valor de eps mas pequeo.
