```python
"""
En este ejercicio utilizamos los datos de gapminder (recordar el ejercicio 10 de la Practica 2
para acceder a los datos de gapminder).
(a) Utilizando seaborn.objects, graficar la curva de la expectativa de vida en Argentina en
funcion del ano, completando el siguiente codigo. Sugerencia: recordar de la practica
anterior como filtrar datos de un dataset.
"""
from gapminder import gapminder
import seaborn . objects as so

datosArgentina = gapminder[gapminder.country == 'Argentina']
#print(datosArgentina)

(
so . Plot ( data = datosArgentina , x = "year", y = "lifeExp")
. add ( so . Line () )
)
```

```python
"""
(b) Realizar un nuevo grafico donde puedan verse las curvas de la expectativa de vida de los
pases de America en funcion del ano, una curva por cada pas.
Sugerencia: utilizar los parametros group = ??? o color = ???. Cual es la diferencia
entre los dos?
"""
americas = gapminder[gapminder.continent == 'Americas']
#display(americas)
(
    so.Plot(data = americas, x = 'year', y = 'lifeExp')
    . add ( so . Line (), color = 'country' )
)
#se diferencia en la visibilidad en este caso
```

```python

"""
(c) Queremos agregar al grafico del tem anterior una curva de tendencia lineal utilizando el
metodo .add(so.Line(), so.PolyFit(1)). Cual de las siguientes dos formas de agrupar
los datos es la forma correcta? Explicar la diferencia entre los dos codigos.
"""

# Codigo 1
(
so . Plot ( data = americas , x ='year' , y ='lifeExp' , group = 'country')
. add ( so . Lines ( color ="#bbca") )
. add ( so . Line () , so . PolyFit (1) )
)
# Codigo 2
(
so . Plot ( data = americas , x ='year' , y ='lifeExp' , group = 'country')
. add ( so . Lines ( color ="#bbca") , group = 'country')
. add ( so . Line () , so . PolyFit (1) )
)

```

```python
"""
(d) Realizar el siguiente grafico, con las curvas de expectativa de vida agrupadas por continente. Sugerencias: que hace el metodo facet() de seaborn.objects.Plot()? Y el
parametro wrap = ??? de facet()?
"""

#facet() permite hacer una plantilla de varios graficos, "wrap=" permite controlar la cantidad de columnas de grficos mostradas
(
    so.Plot(data = gapminder, x = 'year', y = 'lifeExp', group='continent')
    .add(so . Line (), color='country')
).facet(wrap=5)

## FALTA TERMINAR
```
